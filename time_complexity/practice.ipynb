1: Nested Loop Analysis

```python
n = 5
for i in range(n):
    for j in range(i, n):
        print(i,j)


# time complexity = 0(n2)

```

2: Mixed growth rate

```python
def func(n):
    i = 1
    while i < n:
        for j in range(n):
            print(i,j)
        i *= 2


print(func(5))
```

3: Recursive complexity

```python
def solve(n):
    if n <=1:
        return
    solve(n // 2)
    solve(n // 2)
print(solve(1000))
```
