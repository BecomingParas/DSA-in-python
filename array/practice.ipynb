1: Maximum subarray sum(kadane's alogorithms)


```python
def max_subarray(arr):
    max_sum = curr_sum = arr[0]
    for num in arr[1:]:
        curr_sum = max(num, curr_sum + num)
        max_sum = max(max_sum,curr_sum)
    return max_sum

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

print(max_subarray(arr))

```
