Program for sum of n natural numbers
Last Updated : 9 Jan, 2026
Given a positive integer n, find the sum of the first n natural numbers.

Examples : 

Input: n = 3
Output: 6
Explanation: 1 + 2 + 3 = 6

Input: n = 5
Output: 15 
Explanation:  1 + 2 + 3 + 4 + 5 = 15

1. 
[Naive Approach] Using Loop - O(n) Time and O(1) Space
 Calculate the sum of all integers from 1 to n by iterating through a loop.

```python
def find_sum(n):
    sum = 0
    i = 1
    while i <= n:
        sum += i
        i += 1

    return sum

if __name__ == "__main__":
    n = 20
    print(find_sum(20))
```

2. [Alternative Approach] Using Recursion -O(n) and O(n) Space
In this approach, we use recursion to find the sum of the first n natural numbers. The function calls itself with (n-1) until it reaches the base case of n = 1. Each call adds the current value of n to the sum of smaller values, effectively building the result in a top-down manner. 






```python
def find_sum(n):
    if n == 1:
        return 1
    return n + find_sum(n - 1)

if __name__ == '__main__':
    n = 10
    print(find_sum(10))
```

3. [Expected Approach] Formula Based Method- O(1) Time and O(1) Space
Sum of first n natural numbers = (n * (n+1)) / 2

For example: n = 5
Sum = (5 * (5 + 1)) / 2 = (5 * 6) / 2 = 30 / 2 = 15

```python
def find_sum(n):
    return n * (n + 1) // 2

if __name__ == '__main__':
    n = 100
    print(find_sum(n))
```
