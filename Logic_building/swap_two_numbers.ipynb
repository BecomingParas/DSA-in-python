Given two numbers a and b, the task is to swap them.

Examples:

Input: a = 2, b = 3
Output: a = 3, b = 2

Input: a = 20, b = 0
Output: a = 0, b = 20

Input: a = 10, b = 10
Output: a = 10, b = 10 

1. [Naive Approach] Using Third Variable
The idea is to use a third variable, say temp to store the original value of one of the variables during the swap.

Store the value of a in temp.
Assign the value of b to a.
Assign the value of temp to b.

```python
if __name__ == '__main__':
    a = 2
    b = 3
    print("a =", a, "b =", b)
    temp = a
    a = b
    b = temp
    print("a =",a, "b =", b)

# time complexity O(1)
```

2. [Expected Approach] Without using Third Variable
Using Arithmetic Operators
Store the sum of a and b in a (a = a + b).
Get the original value of a, that is (sum - original value of b)and store it in b (b = a - b).
Get the original value of b, that is (sum - original value of a)and store it in a (a = a - b).





```python

if __name__ == "__main__":
    a = 2
    b = 3
    print("a =",a, "b =", b)
    a = a + b
    b = a - b
    a = a - b
    print("a =", a , "b =", b)
```

3. [Alternate Approach] Built-in Swap  tuple unpacking in Python,

```python
if __name__ == '__main__':
    a = 2
    b = 3
    print("a =",a, "b =", b)

    # tuple unpacking
    a, b = b, a
    print("a =",a, "b =", b)
```
