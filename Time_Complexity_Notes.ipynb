## 1. What is Time Complexity?
Time Complexity measures how the execution time of an algorithm grows as input size increases.

---

## 2. Constant Time ‚Äî O(1)
Operations that do not depend on input size.

Examples:
- Array index access
- HashMap lookup

---

## 3. Linear Time ‚Äî O(n)
Execution time grows linearly with input size.

Examples:
- Looping through array
- Linear search

---

## 4. Quadratic Time ‚Äî O(n¬≤)
Nested loops over same input.

Examples:
- Bubble sort
- Selection sort

---

## 5. Cubic Time ‚Äî O(n¬≥)
Triple nested loops.

Examples:
- Floyd Warshall

---

## 6. Logarithmic Time ‚Äî O(log n)
Input size reduced by factor each iteration.

Examples:
- Binary search

---

## 7. Linearithmic Time ‚Äî O(n log n)
Efficient sorting algorithms.

Examples:
- Merge sort
- Quick sort (average)
