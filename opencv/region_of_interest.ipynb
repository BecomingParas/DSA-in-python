```python
import cv2

image = cv2.imread("perfect1.PNG", cv2.IMREAD_COLOR)

print(image)

h, w = image.shape[:2]

print("Height = {}, Width = {}".format(h,w))
```

```python
(B,G,R) = image[100,100]


print("R = {}, G={}, B= {}".format(R,G,B))
```

```python
B = image[100,100,0]

print("B = {}".format(B))
```

```python
roi = image[100: 500, 200: 700]

cv2.imshow("ROI",roi)
cv2.waitKey(0)
```

```python
resize = cv2.resize(image, (1000,1000))

cv2.imshow("Resized image", resize)

cv2.waitKey(0)
```

```python
ratio = 800 / w

dim = (800, int(h * ratio))

resize_aspect = cv2.resize(image, dim)

cv2.imshow('Resized image', resize_aspect)

cv2.waitKey(0)
```

```python
output = image.copy()

rectangle = cv2.rectangle(output, (1500,900),(600,400),(255,0,0),2)

text = cv2.putText(output, 'Opencv Demo', (500,550),cv2.FONT_HERSHEY_SIMPLEX, 4, (255,0,0),2)
```
