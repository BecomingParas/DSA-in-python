```python
import cv2

input_filename = input("Enter the input image file name")
grayscale_image = cv2.imread(input_filename,cv2.IMREAD_GRAYSCALE)
```

```python
if grayscale_image is None:
    print("Error could not load the image")

else:
    cv2.imshow("Grayscale image", grayscale_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    output_filename = input('Enter the output file name to save')
    cv2.imwrite(output_filename, grayscale_image)
    print(f"Grayscale image saved as {output_filename}")
```
